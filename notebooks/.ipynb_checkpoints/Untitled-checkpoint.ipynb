```python
import pandas as pd
import os
```

```python
raw_folder = r"E:\bluestock_mf_capstone\data\raw"
processed_folder = r"E:\bluestock_mf_capstone\data\processed"

files = os.listdir(raw_folder)

print(files)
```

```python
for file in files:
    if file.endswith(".csv"):
        print("Processing:", file)
        df = pd.read_csv(os.path.join(raw_folder, file))

        df = df.drop_duplicates()
        df = df.ffill()
        for column in df.columns:
            if "date" in column.lower():
                try:
                    df[column] = pd.to_datetime(df[column])
                except:
                    pass
        new_name = file.replace(".csv", "_clean.csv")

        df.to_csv(os.path.join(processed_folder, new_name),index=False)
        print(new_name, "Saved Successfully")
```
